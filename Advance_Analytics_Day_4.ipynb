# Covariance

```python
# Covariance for a and b
((df["a"] - df["a"].mean()) * (df["b"] - df["b"].mean())).sum() / (df.shape[0] - 1)
```

```python
#Using numpy to find Covariance
np.cov(df["a"],df["b"])
```

```python
np.cov(df["b"],df["c"])
```

```python
def covariance(x,y):
    
    #finding the mean of the series x & y
    mean_x = sum(x)/float(len(x))
    mean_y = sum(y)/float(len(y))
    
    #Substracting mean from the individual elements
    sub_x = [i - mean_x for i in x]
    sub_y = [i - mean_y for i in y]
    
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    denominator = len(x)-1
    
    cov = numerator/denominator
    return cov

x = df["a"]
y = df["b"]

covariance(x,y)
```
