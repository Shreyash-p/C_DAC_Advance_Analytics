```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statistics
```

```python
iris=pd.read_csv(r"C:\Users\patil\Desktop\Know IT CADC\Adv Analytics\data set\iris.csv")
iris
```

```python
count=iris["Species"].value_counts()
count
```

```python
orders=pd.read_csv(r"C:\Users\patil\Desktop\Know IT CADC\Adv Analytics\data set\Orders.csv")
orders
```

```python
cts=orders["Payment Terms"].value_counts()
cts.plot(kind="bar",color=["red","orange","blue"])
plt.xticks(rotation=45)
plt.show()
```
