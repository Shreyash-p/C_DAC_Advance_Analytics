# Q) find the outliers on Min.Price

```python
# IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP IMP
def detect_outliers(data):
    data = sorted(data)
    q1 = np.percentile(data , 25)
    q3 = np.percentile(data , 75)
    #print(q1, q3)
    IQR = q3 - q1
    lwr_bound = q1 - (1.5 * IQR)
    upr_bound = q3 + (1.5 * IQR)
    #print(lwr_bound , upr_bound)
    outliers = []
    for i in data:
        if (i < lwr_bound or i > upr_bound):
            outliers.append(i)
    return outliers #Driver code
```

```python
sample_outliers = detect_outliers(cars["Min.Price"])
print("Outliers from Min.Price IQR method : ", sample_outliers)

sample_outliers = detect_outliers(cars["Max.Price"])
print("Outliers from Max.Price IQR method : ", sample_outliers)
```
