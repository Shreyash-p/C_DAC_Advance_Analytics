```python
import numpy as np
import pandas as pd
from scipy import stats as stats
from scipy.stats import ttest_1samp, ttest_rel, ttest_ind, bartlett
```

# A group of seven patients of rheumatic heart disease with distention of abdomen due to ascites, affecting breathing capacity were treated. Can we say that treatment has improved breathing capacity? Data is in file Rheumatic.csv

```python
rhuem=pd.read_csv(r"C:\Users\patil\Desktop\Know IT CADC\Adv Analytics\Datasets\Rheumatic.csv")
rhuem
```

```python
# If (Before - After) is < 0
# Then breathing capacity has improved

result = ttest_rel(rhuem['Before'], rhuem['After'], alternative='less')
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)

# Conclusion: We reject H0 at 5% level of significance
# There may be an improvement in breathing capacity
```

# A reacher wants to test whether there is a significant difference in the mean weight of two groups of participants .Perform two-sample t-test at a significant level of 0.05

```python
weight1 = [150,155,160,145,148,152,158,162,155,160,150,153,147,155,149,146,151,159,154,157]
weight2 = [140,145,150,142,135,148,152,142,155,140,142,138,160,145,152,158,156,147,160,161]
```

```python
result = ttest_rel(weight1 , weight2 )
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)

# Conclusion: We reject H0 at 5% level of significance
# There is a significant difference in the mean weight of two groups
```

### Another way

```python
df = {'group1':weight1 , 'group2':weight2}
```

```python
new_df = pd.DataFrame(df)
new_df.head()
```

```python
new_df.to_csv('weight_df.csv' , index=False)
```

```python
# H0 = Their is no significance difference
# H1 = Their is a significance difference
```

```python
result = ttest_rel(new_df['group1'] , new_df['group2'] , alternative='two-sided')
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)

# Conclusion: We reject H0 at 5% level of significance
# There is a significant difference in the mean weight of two groups
```

```python
# Their is no big difference if we use independence(ind)
result = ttest_ind(new_df['group1'] , new_df['group2'] , alternative='two-sided')
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)
```

```python
co2=pd.read_csv(r"C:\Users\patil\Desktop\Know IT CADC\Adv Analytics\Datasets\CO2.csv")
co2
```

```python
chilled = co2[co2['Treatment']=='chilled']
nonchilled = co2[co2['Treatment']=='nonchilled']
```

```python
chilled
```

```python
# H0 = Variance equal
# H1 = Variance unequal
```

```python
result = bartlett(chilled['uptake'] , nonchilled['uptake'])
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)

# Conclusion: We do not reject H0 at 5%
# Variance may be equal
```

```python
# H0 = Mean equal
# H1 = Mean unequal
```

```python
result = ttest_ind(chilled['uptake'] , nonchilled['uptake'] , equal_var=True)
test_stat = result[0]
p_value = result[1]

print("Test Statistics =" , test_stat)
print("P-value =" , p_value)

# Conclusion: We reject H0 at 5%
# Mean may be different
```
